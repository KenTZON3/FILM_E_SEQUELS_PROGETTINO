```python
# 1) Qual è la relazione tra il budget e gli incassi per i film e i loro sequel?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creazione dello scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=mp, x='Budget', y='Incassi', label="film", alpha=0.7)

# Aggiunta di una linea di regressione
sns.regplot(data=mp, x='Budget', y='Incassi', scatter=False, color='red', label='Linea di Regressione')

# Aggiunta di titoli e etichette
plt.title('Correlazione tra Budget e Incassi')
plt.xlabel('Budget (in milioni)')
plt.ylabel('Incassi (in milioni)')
plt.xscale('log')  # Usa la scala logaritmica se i valori sono molto variabili
plt.yscale('log')  # Usa la scala logaritmica se i valori sono molto variabili
plt.legend(title='Legenda')
# Mostra il plot
plt.grid(True)
plt.show()

# RISPOSTA: all'aumentare del budget aumentano anche gli incassi, pochi casi isolati identificano una correlazione negativa per cui un budget più grande non porta necessariamente a maggiori incassi
```

```python
# 2) C'è una correlazione tra il numero di voti su IMDb e il successo al botteghino?

# Creazione dello scatter plot
plt.figure(figsize=(10, 6))

# Plot scatter e linea di regressione
sns.scatterplot(data=mp, x='Num Votes', y='Incassi', alpha=0.7, label='Voti', color='green')
sns.regplot(data=mp, x='Num Votes', y='Incassi', scatter=False, color='red', label='Linea di Regressione')

# Aggiunta di titoli e etichette
plt.title('Correlazione tra Numero di Voti su IMDb e Incassi')
plt.xlabel('Numero di Voti su IMDb')
plt.ylabel('Incassi (in milioni)')

# Imposta la scala logaritmica se necessario
plt.xscale('log')  # Usa la scala logaritmica se i valori sono molto variabili
plt.yscale('log')  # Usa la scala logaritmica se i valori sono molto variabili

# Aggiunta della legenda
plt.legend(title='Legenda')

# Mostra il plot
plt.grid(True)
plt.show()

# Film con un gran numero di voti tendono ad avere maggiori incassi, pochi outlier identificano film con pochi voti ma comunque elevati incassi (correlazione scarsa)
```

```python
# 3) Quali generi di film sono più redditizi?

# Filtriamo i film originali
film_originali = mp[mp['Order'] == 1]

# Calcoliamo il guadagno medio per ciascun genere
# Creiamo un DataFrame con i guadagni per ciascun genere
generi = ['Action', 'Crime', 'Thriller', 'Comedy', 'Adventure', 'Sci-fi', 'Drama', 'Fantasy', 'Sport', 'War', 'Horror', 'Animation', 'Short', 'Mistery', 'Music', 'Romance', 'Family']
guadagni_generi = {genere: 0 for genere in generi}
conteggi_generi = {genere: 0 for genere in generi}

for genere in generi:
    # Selezioniamo i film che appartengono al genere
    film_genere = film_originali[film_originali[genere] == 1]
    # Calcoliamo il guadagno medio per questo genere
    if not film_genere.empty:
        guadagni_generi[genere] = film_genere['Guadagni/Perdite'].mean()
        conteggi_generi[genere] = len(film_genere)

# Convertiamo i risultati in un DataFrame per una visualizzazione più semplice
guadagni_generi_df = pd.DataFrame(list(guadagni_generi.items()), columns=['Genere', 'Guadagno Medio'])
guadagni_generi_df = guadagni_generi_df.sort_values(by='Guadagno Medio', ascending=False)

# Creiamo un grafico a barre
plt.figure(figsize=(12, 8))
bars = plt.barh(guadagni_generi_df['Genere'], guadagni_generi_df['Guadagno Medio'], color='skyblue')

# Aggiungiamo etichette sui valori delle barre
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{width:.2f}', va='center')

# Aggiungiamo le etichette e il titolo
plt.xlabel('Guadagno Medio (in milioni)')
plt.ylabel('Genere')
plt.title('Guadagno medio per genere nei film originali')

# Inverti l'asse Y per avere i generi con film originali più redditizi in cima
plt.gca().invert_yaxis()

# Mostriamo il grafico
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# domina il genere Animazione (1), Musicale (2), Avventura ((3)
```

```python
# Quali generi tendono a produrre i sequel più redditizi?

# Filtriamo i sequel
sequel = mp[mp['Order'] > 1]

# Calcoliamo il guadagno medio per ciascun genere dei sequel
# Creiamo un DataFrame con i guadagni per ciascun genere
generi = ['Action', 'Crime', 'Thriller', 'Comedy', 'Adventure', 'Sci-fi', 'Drama', 'Fantasy', 'Sport', 'War', 'Horror', 'Animation', 'Short', 'Mistery', 'Music', 'Romance', 'Family']
guadagni_sequel_generi = {genere: 0 for genere in generi}
conteggi_sequel_generi = {genere: 0 for genere in generi}

for genere in generi:
    # Selezioniamo i sequel che appartengono al genere
    sequel_genere = sequel[sequel[genere] == 1]
    # Calcoliamo il guadagno medio per questo genere
    if not sequel_genere.empty:
        guadagni_sequel_generi[genere] = sequel_genere['Guadagni/Perdite'].mean()
        conteggi_sequel_generi[genere] = len(sequel_genere)

# Convertiamo i risultati in un DataFrame per una visualizzazione più semplice
guadagni_sequel_generi_df = pd.DataFrame(list(guadagni_sequel_generi.items()), columns=['Genere', 'Guadagno Medio'])
guadagni_sequel_generi_df = guadagni_sequel_generi_df.sort_values(by='Guadagno Medio', ascending=False)

# Creiamo un grafico a barre orizzontali
plt.figure(figsize=(12, 8))
bars = plt.barh(guadagni_sequel_generi_df['Genere'], guadagni_sequel_generi_df['Guadagno Medio'], color='lightcoral')

# Aggiungiamo etichette sui valori delle barre
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{width:.2f}', va='center')

# Aggiungiamo le etichette e il titolo
plt.xlabel('Guadagno Medio (in milioni)')
plt.ylabel('Genere')
plt.title('Guadagno medio dei sequel per genere')

plt.gca().invert_yaxis()

# Mostriamo il grafico
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# domina il genere Animazione (1), Fantasy (2), Adventure (3)
```

```python
# GRAFICI UNITI

# Filtriamo i film originali e i sequel
film_originali = mp[mp['Order'] == 1]
sequel = mp[mp['Order'] > 1]

# Calcoliamo il guadagno medio per ciascun genere
generi = ['Action', 'Crime', 'Thriller', 'Comedy', 'Adventure', 'Sci-fi', 'Drama', 'Fantasy', 'Sport', 'War', 'Horror', 'Animation', 'Short', 'Mistery', 'Music', 'Romance', 'Family']

# Guadagni medi per film originali
guadagni_originali_generi = {genere: 0 for genere in generi}
for genere in generi:
    film_genere = film_originali[film_originali[genere] == 1]
    if not film_genere.empty:
        guadagni_originali_generi[genere] = film_genere['Guadagni/Perdite'].mean()

# Guadagni medi per sequel
guadagni_sequel_generi = {genere: 0 for genere in generi}
for genere in generi:
    sequel_genere = sequel[sequel[genere] == 1]
    if not sequel_genere.empty:
        guadagni_sequel_generi[genere] = sequel_genere['Guadagni/Perdite'].mean()

# Convertiamo i risultati in un DataFrame per una visualizzazione più semplice
guadagni_originali_generi_df = pd.DataFrame(list(guadagni_originali_generi.items()), columns=['Genere', 'Guadagno Medio Originali'])
guadagni_sequel_generi_df = pd.DataFrame(list(guadagni_sequel_generi.items()), columns=['Genere', 'Guadagno Medio Sequel'])

# Uniamo i DataFrame
combined_df = pd.merge(guadagni_originali_generi_df, guadagni_sequel_generi_df, on='Genere')

# Ordinamento per Guadagno Medio Originali (puoi scegliere di ordinare come preferisci)
combined_df = combined_df.sort_values(by='Guadagno Medio Originali', ascending=False)

# Creiamo un grafico a barre affiancate
plt.figure(figsize=(14, 10))
bar_width = 0.4  # Larghezza delle barre
index = np.arange(len(combined_df['Genere']))

# Barre dei guadagni medi dei film originali
bars1 = plt.barh(index - bar_width/2, combined_df['Guadagno Medio Originali'], bar_width, label='Film Originali', color='skyblue')

# Barre dei guadagni medi dei sequel
bars2 = plt.barh(index + bar_width/2, combined_df['Guadagno Medio Sequel'], bar_width, label='Sequel', color='lightcoral')

# Aggiungiamo etichette sui valori delle barre
for bar in bars1:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{width:.2f}', va='center')

for bar in bars2:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{width:.2f}', va='center')

# Aggiungiamo le etichette e il titolo
plt.xlabel('Guadagno Medio (in milioni)')
plt.ylabel('Genere')
plt.title('Confronto dei Guadagni Medi per Genere: Film Originali vs Sequel')
plt.yticks(index, combined_df['Genere'])
plt.legend()

plt.gca().invert_yaxis()

# Mostriamo il grafico
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# I guadagni maggiori provengono dai sequel, eccetto per il genere musicale, la commedia, horror e guerra
```

```python
# Quali sono i generi più longevi in termini di numero di sequel?
import pandas as pd
import matplotlib.pyplot as plt

# Definizione dei generi
generi = ['Action', 'Crime', 'Thriller', 'Comedy', 'Adventure', 'Sci-fi', 'Drama', 'Fantasy',
          'Sport', 'War', 'Horror', 'Animation', 'Short', 'Mistery', 'Music', 'Romance', 'Family']

# Filtro per selezionare solo i sequel (Order > 1)
sequel = mp[mp['Order'] > 1]

# Conta il numero di sequel per ogni genere
sequel_count_per_genere = {}

for genere in generi:
    sequel_count_per_genere[genere] = sequel[sequel[genere] == 1].shape[0]

# Converti il dizionario in DataFrame
df_sequel_count = pd.DataFrame(list(sequel_count_per_genere.items()), columns=['Genere', 'Numero di Sequel'])

# Ordina i generi per numero di sequel
df_sequel_count = df_sequel_count.sort_values(by='Numero di Sequel', ascending=False)

# Creazione del grafico a barre
plt.figure(figsize=(12, 8))
plt.barh(df_sequel_count['Genere'], df_sequel_count['Numero di Sequel'], color='lightblue')
plt.xlabel('Numero di Sequel')
plt.ylabel('Genere')
plt.title('Generi più Longevi in termini di Numero di Sequel')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Inverti l'asse Y per avere i generi con il maggior numero di sequel in cima
plt.gca().invert_yaxis()

# Mostra il grafico
plt.show()

# I generi Action, Adventure, Thriller, Comedy, Horror sono quelli che dominano in termini di sequel ma il loro successo com'è in relazione a ciò?
```

```python
# Definizione dei generi
generi = ['Action', 'Crime', 'Thriller', 'Comedy', 'Adventure', 'Sci-fi', 'Drama', 'Fantasy',
          'Sport', 'War', 'Horror', 'Animation', 'Short', 'Mistery', 'Music', 'Romance', 'Family']

# Filtro per selezionare solo i sequel (Order > 1)
sequel = mp[mp['Order'] > 1]

# Conta il numero di sequel per ogni genere, guadagni medi e percentuale di successi
longevità_per_genere = {}
guadagni_per_genere = {}
successo_flop_per_genere = {}

for genere in generi:
    # Longevità: Numero di sequel per ogni genere
    longevità_per_genere[genere] = sequel[sequel[genere] == 1].shape[0]
    
    # Guadagno medio per genere
    guadagni_per_genere[genere] = sequel[sequel[genere] == 1]['Guadagni/Perdite'].mean()
    
    # Percentuale di successi e flop per genere
    totale_films_genere = sequel[sequel[genere] == 1].shape[0]
    successi = sequel[(sequel[genere] == 1) & (sequel['Risultati'] == 'Successo')].shape[0]
    if totale_films_genere > 0:
        successo_flop_per_genere[genere] = successi / totale_films_genere
    else:
        successo_flop_per_genere[genere] = 0

# Converti in DataFrame per grafico
df_longevo_guadagni = pd.DataFrame({
    'Genere': generi,
    'Longevità (Numero di Sequel)': [longevità_per_genere[genere] for genere in generi],
    'Guadagno Medio': [guadagni_per_genere[genere] for genere in generi],
    'Percentuale Successi': [successo_flop_per_genere[genere] for genere in generi]
})

# Gestione dei valori NaN o infiniti
df_longevo_guadagni.replace([np.inf, -np.inf], np.nan, inplace=True)
df_longevo_guadagni.dropna(subset=['Longevità (Numero di Sequel)', 'Guadagno Medio'], inplace=True)

# Creazione del grafico a dispersione (longevità vs guadagni medi)
plt.figure(figsize=(10, 6))

# Scatter plot per Longevità vs Guadagno Medio
scatter = plt.scatter(df_longevo_guadagni['Longevità (Numero di Sequel)'], 
                      df_longevo_guadagni['Guadagno Medio'], 
                      c=df_longevo_guadagni['Percentuale Successi'], cmap='coolwarm', s=100, alpha=0.8)

# Etichette per i punti
for i, genere in enumerate(df_longevo_guadagni['Genere']):
    plt.text(df_longevo_guadagni['Longevità (Numero di Sequel)'].iloc[i] + 0.1, 
             df_longevo_guadagni['Guadagno Medio'].iloc[i] + 0.1, 
             genere, fontsize=9)

# Etichette e titoli
plt.xlabel('Longevità (Numero di Sequel)')
plt.ylabel('Guadagno Medio (in milioni)')
plt.title('Longevità vs Guadagno Medio per Genere')

# Aggiungi la barra dei colori (Colorbar) per rappresentare la percentuale di successi
plt.colorbar(scatter, label='Percentuale di Successi')

# Mostra il grafico
plt.grid(True)
plt.show()

# Maggiore la longevità in termini di sequel non è direttamente proporzionale a maggiori guadagni, ad esempio il genere Animazione
# ha molti guadagni ma offre meno sequels
# il genere avventura è quello più bilanciato (tanti sequel e buoni incassi)
```

```python
# Che relazione c'è tra i guadagni dei film originali e sequel negli anni? 

from scipy.stats import pearsonr

# Filtriamo i film originali e i sequel
film_originali = mp[mp['Order'] == 1]
sequel = mp[mp['Order'] > 1]

# Filtriamo gli anni zero
mp = mp[mp['Year'] > 0]

# Troviamo l'anno minimo di rilascio nel dataset, escludendo gli anni zero
anno_minimo = mp['Year'].min()
print(f"Anno minimo di rilascio (escludendo gli anni zero): {anno_minimo}")

# Calcoliamo i guadagni medi annuali per film originali e sequel
guadagni_originali_annuali = film_originali.groupby('Year')['Guadagni/Perdite'].mean().reset_index()
guadagni_sequel_annuali = sequel.groupby('Year')['Guadagni/Perdite'].mean().reset_index()

# Calcoliamo il guadagno medio per film originali e sequel
guadagno_medio_originali = guadagni_originali_annuali['Guadagni/Perdite'].mean()
guadagno_medio_sequel = guadagni_sequel_annuali['Guadagni/Perdite'].mean()
print(f"Guadagno medio film originali: {guadagno_medio_originali}")
print(f"Guadagno medio sequel: {guadagno_medio_sequel}")

# Filtriamo per iniziare dall'anno minimo di rilascio
guadagni_originali_annuali = guadagni_originali_annuali[guadagni_originali_annuali['Year'] >= anno_minimo]
guadagni_sequel_annuali = guadagni_sequel_annuali[guadagni_sequel_annuali['Year'] >= anno_minimo]

# Calcoliamo il coefficiente di correlazione di Pearson tra i guadagni dei film originali e dei sequel
merged_data = pd.merge(guadagni_originali_annuali, guadagni_sequel_annuali, on='Year', suffixes=('_originali', '_sequel'))
correlation, _ = pearsonr(merged_data['Guadagni/Perdite_originali'], merged_data['Guadagni/Perdite_sequel'])
print(f"Coefficiente di Correlazione di Pearson: {correlation:.2f}")

# Creiamo un grafico per visualizzare i guadagni annuali dei film originali e dei sequel
plt.figure(figsize=(12, 6))

# Grafico dei guadagni medi annuali dei film originali
plt.plot(guadagni_originali_annuali['Year'], guadagni_originali_annuali['Guadagni/Perdite'], label='Film Originali', marker='o', color='blue')

# Grafico dei guadagni medi annuali dei sequel
plt.plot(guadagni_sequel_annuali['Year'], guadagni_sequel_annuali['Guadagni/Perdite'], label='Sequel', marker='o', color='green')

# Aggiungiamo le linee del guadagno medio
plt.axhline(y=guadagno_medio_originali, color='blue', linestyle='--', label=f'Guadagno Medio Film Originali: {guadagno_medio_originali:.2f}')
plt.axhline(y=guadagno_medio_sequel, color='green', linestyle='--', label=f'Guadagno Medio Sequel: {guadagno_medio_sequel:.2f}')

# Aggiungiamo il coefficiente di correlazione al grafico
plt.text(anno_minimo + (guadagni_originali_annuali['Year'].max() - anno_minimo) * 0.05,
         max(guadagni_originali_annuali['Guadagni/Perdite'].max(), guadagni_sequel_annuali['Guadagni/Perdite'].max()) * 0.9,
         f'Correlazione: {correlation:.2f}', fontsize=12, color='black')

# Aggiungiamo le etichette e il titolo
plt.xlabel('Anno')
plt.ylabel('Guadagni Medi (in milioni)')
plt.title('Relazione tra Guadagni dei Film Originali e Sequel negli Anni')
plt.legend(loc='upper center')

plt.axvline(x=1942)
plt.axvline(x=2022)
plt.axvspan(1975,1980, facecolor='orange', alpha=0.4)
plt.axvspan(1989,1996, facecolor='orange', alpha=0.4)
plt.axvspan(2015,2022, facecolor='orange',alpha=0.4)

# Imposta i limiti dell'asse X per partire dall'anno minimo
plt.xlim(left=anno_minimo)

# Mostriamo il grafico
plt.grid(True)
plt.show()

# bassa correlazione, fascie arancione: 1975-1980 più successo film originali; 1989-1996 più successo film originali; 2015-2022 più successo i sequel
```
